```python
import os
import numpy as np
from shutil import copyfile

Path = "/home/ug-ml/felix-ML/VAE_000/Data/Data"

```

```python
def Normalise(Image): #Shape N by N
    min = Image[0][0]
    max = Image[0][0]
    for i in Image:
        for j in i:
            if(j > max):
                max = j
            if(j < min):
                min = j
    #print(min, max)
    Image = (Image - min) / (max - min)
    return Image
```

```python
Path_i = sorted(os.listdir(Path)) #Training Validation and Test
for i in Path_i: #i training, validation and test
    Path_j = sorted(os.listdir(Path + "/" + i))
    for j in Path_j: #j is the crystal
        InputFile = Path + "/" + i + "/" + j +"/" + "Input.npy"
        OutputFile = Path + "/" + i + "/" + j +"/" + "Output.npy"
        Image_Input = np.load(InputFile).astype(np.float32)
        Image_Output = np.load(OutputFile).astype(np.float32)
        
        
        Image_Input = Normalise(Image_Input)
        Image_Output = Normalise(Image_Output)
        np.save(InputFile, Image_Input)
        np.save(OutputFile, Image_Output)
        print(j)
```
